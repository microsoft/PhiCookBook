# Guidance-യെക്കുറിച്ച്
Guidance ഒരു തെളിയിച്ച ഓപ്പൺ-സോഴ്സ് Python ലൈബ്രറിയാണ്, ഏതെങ്കിലും ഭാഷാ മാതൃകയുടെ (LM) ഔട്ട്പുട്ടുകൾ നിയന്ത്രിക്കാൻ. ഒരൊറ്റ API കോൾ ഉപയോഗിച്ച് (Python-ൽ) മോഡൽ പാലിക്കേണ്ട കൃത്യമായ പ്രോഗ്രാമാറ്റിക് നിയന്ത്രണ(ങ്ങൾ) നിങ്ങൾ വ്യക്തമാക്കാം, അതിലൂടെ ആവശ്യമായ ഘടനാപരമായ ഔട്ട്‌പുട്ടുകൾ JSON, Python, HTML, SQL, അല്ലെങ്കിൽ ഉപയോഗകേസിന് ആവശ്യമായ ഏതെങ്കിലും ഘടനയിൽ ജനറേറ്റ് ചെയ്യാൻ കഴിയും.

Guidance പരമ്പരാഗത പ്രോംപ്റ്റിംഗ് സാങ്കേതികവിദ്യകളിൽ നിന്ന് വ്യത്യസ്തമാണ്. ഇത് ഇൻഫെറൻസ് ലെയറിൽ മോഡൽ ഓരോ ടോക്കണിനും നയിച്ച് നിയന്ത്രണങ്ങൾ നടപ്പിലാക്കുന്നു, അതിലൂടെ ഉയർന്ന ഗുണനിലവാരമുള്ള ഔട്ട്‌പുട്ടുകൾ നൽകുകയും,高度ഘടനാപരമായ സാഹചര്യങ്ങളിൽ ഉപയോഗിക്കുമ്പോൾ ചെലവും വൈകല്യവും 30–50% വരെ കുറയ്ക്കുകയും ചെയ്യുന്നു.

Guidance-യെക്കുറിച്ച് കൂടുതൽ അറിയാൻ, [GitHub-ലുള്ള പൊതുജന റിപോസിറ്ററി](https://github.com/guidance-ai/guidance) സന്ദർശിക്കുക അല്ലെങ്കിൽ Microsoft Build-ൽ നടന്ന [Guidance ബ്രേക്കൗട്ട് സെഷൻ](https://www.youtube.com/watch?v=qXMNPVVlCMs) കാണുക.


# ക്രമീകരണം
1. Guidance ഇൻസ്റ്റാൾ ചെയ്യുക `pip install guidance --pre` ഉപയോഗിച്ച്
2. Azure-ൽ Phi 3.5 mini endpoint വിന്യസിക്കാൻ https://ai.azure.com/explore/models/Phi-3.5-mini-instruct/version/2/registry/azureml സന്ദർശിച്ച് "Deploy" ബട്ടണിൽ ക്ലിക്ക് ചെയ്യുക
3. നിങ്ങളുടെ endpoint-ന്റെ API കീ `AZURE_PHI3_KEY` എന്ന environment variable-ൽ സംഭരിക്കുകയും URL `AZURE_PHI3_URL` എന്ന environment variable-ൽ സംഭരുകയും ചെയ്യുക


```python
from guidance import gen, select, regex, user, assistant, system, json
from guidance.models import AzureGuidance
from json import loads as load_json_str
import os

phi3_url = os.getenv("AZURE_PHI3_URL")
phi3_api_key = os.getenv("AZURE_PHI3_KEY")
phi3_lm = AzureGuidance(f"{phi3_url}/guidance#auth={phi3_api_key}")

# Or, load from HuggingFace to run locally
# from guidance.models import Transformers
# phi3_lm = Transformers("microsoft/Phi-3-mini-4k-instruct")
```

# നിയന്ത്രണരഹിത സൃഷ്ടി
`gen()` ഫംഗ്ഷൻ ഉപയോഗിച്ച് ഏതെങ്കിലും നിയന്ത്രണങ്ങളില്ലാതെ ടെക്സ്റ്റ് സൃഷ്ടിക്കാവുന്നതാണ്. ഇത് Guidance ഇല്ലാതെ മോഡൽ ഉപയോഗിക്കുന്നതിനെ സമാനമാണ്.

## ചാറ്റ് ഫോർമാറ്റിംഗ്
ബഹുഭൂരിപക്ഷം ചാറ്റ് മോഡലുകളെപ്പോലെ, Phi-3 ഉപയോക്താവിനും അസിസ്റ്റന്റിനും മധ്യസ്ഥമായി ഉണ്ടായിരിക്കുന്ന സന്ദേശങ്ങളെ ഒരു പ്രത്യേക ഫോർമാറ്റിൽ പ്രതീക്ഷിക്കുന്നു. Guidance Phi-3യുടെ ചാറ്റ് ടെംപ്ലേറ്റ് പിന്തുണയ്ക്കുകയും നിങ്ങള്‍ക്കായി ചാറ്റ് ഫോർമാറ്റിംഗ് കൈകാര്യം ചെയ്യുകയും ചെയ്യും. ചാറ്റ് ടേണുകൾ സൃഷ്ടിക്കാൻ, സംഭാഷണത്തിലെ ഓരോ ഭാഗവും `with user()` അല്ലെങ്കിൽ `with assistant()` ബ്ലോക്കുകളിലിടുക. ഒരു `with system()` ബ്ലോക്ക് സിസ്റ്റം മെസ്സേജ് സജ്ജമാക്കാൻ ഉപയോഗിക്കാവുന്നതാണ്.


```python
lm = phi3_lm
with system():
    lm += "You are a helpful assistant. You have a cranky yet entertaining temperament."
with user():
    lm += "What is the capital of Australia?"
with assistant():
    lm += gen(temperature=0.8, max_tokens=100)
```

## ടോക്കൺ ലാഭം

ഉയർന്ന ഘടനയുള്ള സാഹചര്യങ്ങളിൽ, Guidance ടോക്കൺಗಳು ഒഴിവാക്കി ആവശ്യമായ ടോക്കൺകൾ മാത്രമേ സൃഷ്ടിക്കൂ, ഇതിലൂടെ പ്രവർത്തനക്ഷമത മെച്ചപ്പെടുകയും കാര്യക്ഷമത വർദ്ധിക്കുകയും API ചെലവുകൾ ലാഭിക്കുകയും ചെയ്യുന്നു. സൃഷ്ടിച്ച ടോക്കൺകൾ ഈ നോട്ട്ബുക്കിൽ ഹൈലൈറ്റ് ചെയ്ത പശ്ചാത്തലത്തോടെ കാണിക്കുന്നു. ബലികൃത ടോക്കൺകൾ ഹൈലൈറ്റ് ചെയ്യാതെ കാണിക്കുന്നു, കൂടാതെ അവയുടെ ചെലവ് ഇൻപുട്ട് ടോക്കൺകളുടെ ചെലവിനോട് സമമാണ്, അത് ഔട്ട്പുട്ട് ടോക്കൺകളുടെ ചെലവിന്റെ ഏക-മൂന്നിലൊന്ന് എന്ന് കണക്കാക്കപ്പെടുന്നു.

*സൂചനം:* നിയന്ത്രണരഹിതമായ ജനറേഷനുള്ള ആദ്യ ഉദാഹരണം ഏതെങ്കിലും ടോക്കൺകൾ നിർബന്ധിതമാക്കാൻ കഴിഞ്ഞില്ല, കാരണം നമുക്ക് യാതൊരു നിയന്ത്രണവും നൽകിയിരുന്നില്ല.


# Phi 3-ന് വേണ്ടി സംസാരിക്കുക 

Guidance ഉപയോഗിച്ച്, മോഡലിന്റെ പ്രതികരണങ്ങളിൽ നിങ്ങൾ എളുപ്പത്തിൽ ടെക്സ്റ്റ് ചേർക്കാൻ കഴിയും. നിങ്ങൾ മോഡലിന്റെ ഔട്ട്‌പുട്ട് ഒരു പ്രത്യേക ദിശയിൽ നയിക്കാൻ ആഗ്രഹിക്കുകയാണെങ്കിൽ ഇത് സഹായകരമായിരിക്കും.


```python
lm = phi3_lm
with user():
    lm += "What is the capital of Australia?"
with assistant():
    lm += "The capital of Australia is " + gen(temperature=0.8, max_tokens=50)
```

`ഓസ്‌ട്രേലിയയുടെ തലസ്ഥാനം ആണ്` ഹൈലൈറ്റ് ചെയ്യപ്പെട്ടിട്ടില്ല, കാരണം അസിസ്റ്റന്റിന്റെ പ്രതികരണത്തിലെ ആ ഭാഗം Guidance നാൽ നിർബന്ധിതമായതിനാലാണ്.


# regex ഉപയോഗിച്ച് നിയന്ത്രിക്കൽ
മുമ്പത്തെ ഉദാഹരണത്തിൽ, Phi 3 ചോദ്യത്തിന് `Canberra` എന്ന് മറുപടി നൽകിയതിനു ശേഷം പിന്തുടരുന്ന വിശദീകരണങ്ങളുമായി പ്രതികരിച്ചു. മോഡലിന്റെ ഔട്ട്പുട്ട് കൃത്യമായി ഒരു വാക്കായി പരിമിതപ്പെടുത്താൻ ഒരു regex ഉപയോഗിക്കാം.


```python
lm = phi3_lm
with user():
    lm += "What is the capital of Australia?"
with assistant():
    lm += "The capital of Australia is " + regex("[A-Z][a-z]+")
```

regex ഉപയോഗിച്ച്, `Canberra` എന്നതേ മാത്രമാണ് സൃഷ്ടിക്കപ്പെടുന്നത്.


# പല തിരഞ്ഞെടുപ്പ് ഓപ്ഷനുകളിൽ നിന്ന് തിരഞ്ഞെടുക്കൽ
ചില സാദ്ധ്യമായ തിരഞ്ഞെടുപ്പ് ഓപ്ഷനുകൾ അറിയാമെങ്കിൽ, മോഡലിനെ ഓപ്ഷനുകളുടെ ഒരു പട്ടികയിൽ നിന്ന് തിരഞ്ഞെടുക്കാൻ `select()` ഫംഗ്ഷൻ ഉപയോഗിക്കാം.


```python
lm = phi3_lm
with user():
    lm += "What is the capital of Australia?"
with assistant():
    lm += "The capital of Australia is " + select(["Washington", "Canberra", "Sydney", "Melbourne"])
```

`select()` ഉപയോഗിച്ചപ്പോൾ, `Can` എന്ന ടോക്കൺ മാത്രം സൃഷ്ടിക്കപ്പെട്ടു. `Canberra` മാത്രമേ പ്രതികരണം പൂർത്തിയാക്കാൻ സാധിക്കുകയുള്ളൂ എന്നതിനാൽ, ശേഷിക്കുന്ന ടോക്കണുകൾ നിർബന്ധിതമായി ഉണ്ടാക്കിയവയായിരുന്നു.


# چിന്താശൃംഖല

ചിന്താശൃംഖല ഒരു സാങ്കേതിക വിദ്യയാണ്, ഇത് ഒരു പ്രശ്നം ഘട്ടം ഘട്ടമായി പ്രോസസ് ചെയ്യാൻ പ്രേരിപ്പിച്ച് മോഡലിന്റെ ഫലത്തിന്റെ ഗുണമേന്മ മെച്ചപ്പെടുത്താൻ സഹായിക്കുന്നു. സാധാരണയായി, ഒരു അന്തിമ ഉത്തരത്തിൽ എത്താൻ ഒന്നിലധികം പ്രോമ്പ്റ്റ് തവണകൾ ആവശ്യമാകും. ആദ്യം, മോഡലിനെ ഘട്ടം ഘട്ടമായി ആലോചിക്കാൻ നിർദ്ദേശിക്കുക. പിന്നീട്, അന്തിമ ഉത്തരം നൽകാൻ മോഡലിനെ വീണ്ടും പ്രോമ്പ്റ്റ് ചെയ്യുക. സ്റ്റാൻഡാർഡ് ചാറ്റ് ഇൻഫറൻസ് APIs ഉപയോഗിച്ചാൽ, ഇതിന് 2 API കോൾ എടുക്കും, കൂടാതെ മോഡൽ സൃഷ്ടിച്ച “ചിന്താശൃംഖലം” രണ്ട് തവണ ചാർജ് ചെയ്യപ്പെടും – ഒരിക്കൽ മോഡൽ അത് സൃഷ്ടിക്കുമ്പോൾ ഔട്ട്‌പുട്ട് ടോക്കണുകളായി, പിന്നീട് രണ്ടാം കോൾക്കുള്ള ഇൻപുട്ട് ടോക്കണുകളായി. Guidance ഉപയോഗിച്ച്,  മുഴുവൻ ബഹുഘട്ട പ്രക്രിയ ഒരു ഒറ്റ API കോൾ ആയി പ്രോസസ് ചെയ്ത് അതിന്റെ ഭാഗമായാണ് ചാർജ് ചെയ്യപ്പെടുന്നത്, ചെലവും ലേറ്റൻസിയും കുറയുന്നു.


```python
gsm8k_question = "Mark has a garden with flowers. He planted plants of three different colors in it. Ten of them are yellow, and there are 80% more of those in purple. There are only 25% as many green flowers as there are yellow and purple flowers. How many flowers does Mark have in his garden?"
lm = phi3_lm
with user():
    lm += gsm8k_question
with assistant():
    lm += "Let's think step by step. " + gen(temperature=0.8, max_tokens=500)
    # Prompt for the final answer, which should be a number. Store the output in an "answer" variable.
    lm += "\nTherefore, the final answer is: " + regex(r"\d+", name="answer")

print(f"Final answer: {lm['answer']}")
```

# JSON ജനറേഷൻ
ദിശാനിർദേശങ്ങൾ JSON സ്കീമയോ pydantic മോഡലോ അനുസരിച്ച് വരുന്ന JSON ഉത്പാദനം ഉറപ്പാക്കാൻ ഉപയോഗിക്കാം, ഉദാഹരണത്തിന് ഇവിടെ കാണുന്ന ഉപയോക്തൃ പ്രൊഫൈൽ സ്കീമ പോലുള്ളത്.


```python
user_json_schema = load_json_str("""{
  "$schema": "http://json-schema.org/draft-07/schema#",
  "title": "User Profile",
  "type": "object",
  "properties": {
    "username": {
      "type": "string"
    },
    "age": {
      "type": "integer"
    },
    "email": {
      "type": "string"
    }
  },
  "additionalProperties": false
}
""")

lm = phi3_lm
with user():
    lm += "Generate a JSON object for a user profile. The profile should include a username, age, email, and nothing more."

with assistant():
    lm += json(schema=user_json_schema, temperature=1.0)
```

```python
from pydantic import BaseModel

class UserProfile(BaseModel):
    username: str
    age: int
    email: str


lm = phi3_lm
with user():
    lm += "Generate a JSON object for a user profile. The profile should include a username, age, email, and nothing more."

with assistant():
    lm += json(schema=UserProfile, temperature=1.0)
```

## HTML നിര്‍മ്മാണം

Guidance പ്രോഗ്രാമിംഗ് ഭാഷയിലെ വ്യാകരണ ആവശ്യങ്ങൾ പാലിക്കുകയും കോഡ് സൃഷ്ടിക്കുന്നതിനും ഉപയോഗിക്കാം. In this section, we will create a small Guidance program for writing very simple HTML webpages.

നാം വെബ്‌പേജ് ചെറിയ ഭാഗങ്ങളാക്കി വിഭജിക്കുകയും ഓരോ ഭാഗത്തിനും സ്വന്തം Guidance ഫങ്ഷൻ ഉണ്ടാക്കുകയും ചെയ്യും.  അവയെ പിന്നീട് നമുക്ക് അവസാന ഫങ്ഷനിൽ ചേർത്ത് ഒരു HTML വെബ്‌പേജ് സൃഷ്ടിക്കാൻ സംയോജിപ്പിക്കും.

പിന്നീട് നാം ഈ ഫങ്ഷൻ Guidance-സജ്ജമായ Azure AI മോഡൽയിൽ പ്രവർത്തിപ്പിക്കും.

*Note:* ഇത് പൂര്‍ണവിശേഷതകളുള്ള ഒരു HTML ജനറേറ്റർ അല്ല; ലക്ഷ്യം നിങ്ങളുടെ വ്യക്തിഗത ആവശ്യങ്ങൾക്ക് അനുയോജ്യമായി നിങ്ങൾ എങ്ങനെ ഘടനാപരമായ ഔട്ട്പുട്ട് സൃഷ്ടിക്കാവുന്നതാണ് എന്ന് കാണിക്കുക

നാം Guidance-ൽ നിന്നുള്ള ആവശ്യമായതു ഇറക്കുമതി ചെയ്യുന്നതോടെ തുടങ്ങാം:


```python
from guidance import guidance
from guidance.library import (
    zero_or_more,
    any_char_but,
    select,
    capture,
    with_temperature,
)
from guidance.models import Model
```

HTML വെബ്‌പേജുകൾ вельмі ഘടനാപരമാണ്, ഞങ്ങൾ Guidance ഉപയോഗിച്ച് പേജിന്റെ ആ ഭാഗങ്ങൾ 'നിർബന്ധിതമാക്കും'.
When we explicitly require text from the model, we need to ensure it doesn't include anything which could be a tag - that is, we must exclude the '<' and '>' characters:


```python
@guidance(stateless=True)
def _gen_text(lm: Model):
    return lm + zero_or_more(any_char_but(["<", ">"]))
```

പിന്നീട് നാം ഈ ഫംഗ്ഷൻ ഉപയോഗിച്ച് ഏതെങ്കിലും HTML ടാഗിനുള്ളിൽ ടെക്സ്റ്റ് സൃഷ്ടിക്കാം:


```python
@guidance(stateless=True)
def _gen_text_in_tag(lm: Model, tag: str):
    lm += f"<{tag}>"
    lm += _gen_text()
    lm += f"</{tag}>"
    return lm
```

ഇതിൽ, നാം പേജ് ഹെഡർ സൃഷ്‌ടിക്കാം.
ഇതിന്റെ ഭാഗമായായി, നമുക്ക് ഒരു പേജ് ശീർഷകം സൃഷ്‌ടിക്കേണ്ടതാണ്:


```python
@guidance(stateless=True)
def _gen_header(lm: Model):
    lm += "<head>\n"
    lm += _gen_text_in_tag("title") + "\n"
    lm += "</head>\n"
    return lm
```

HTML പേജിന്റെ ബോഡി ഭാഗം ശീർഷകങ്ങളാൽയും പാരഗ്രാഫുകളാലും നിറയാൻ പോകുന്നു.
നമുക്ക് ഓരോന്നും ചെയ്യാൻ ഒരു ഫംഗ്ഷൻ നിർവചിക്കാം:


```python
@guidance(stateless=True)
def _gen_heading(lm: Model):
    lm += select(
        options=[_gen_text_in_tag("h1"), _gen_text_in_tag("h2"), _gen_text_in_tag("h3")]
    )
    lm += "\n"
    return lm

@guidance(stateless=True)
def _gen_para(lm: Model):
    lm += _gen_text_in_tag("p")
    lm += "\n"
    return lm
```

ഇപ്പോൾ, HTML-യുടെ ബോഡി തന്നെ നിർവചിക്കുന്ന ഫംഗ്ഷൻ.
ഇത് `select()`-നെ `recurse=True` ആക്കിയാണ് അനേകം ഹെഡിംഗുകളും പാരഗ്രാഫുകളും സൃഷ്ടിക്കുന്നത്:


```python
@guidance(stateless=True)
def _gen_body(lm: Model):
    lm += "<body>\n"
    lm += select(options=[_gen_heading(), _gen_para()], recurse=True)
    lm += "</body>\n"
    return lm
```

അടുത്തതായി, പൂർണ്ണമായ HTML പേജ് സൃഷ്ടിക്കുന്ന ഫംഗ്ഷനിലേക്ക് നമുക്ക് വരാം.
നാം HTML ആരംഭ ടാഗ് ചേർക്കുന്നു, തുടർന്ന് ഹെഡർ സൃഷ്ടിക്കുന്നു, പിന്നെ ബോഡി സൃഷ്ടിക്കുകയും ശേഷം അവസാന HTML ടാഗ് ചേർക്കുകയും ചെയ്യുന്നു:


```python
@guidance(stateless=True)
def _gen_html(lm: Model):
    lm += "<html>\n"
    lm += _gen_header()
    lm += _gen_body()
    lm += "</html>\n"
    return lm
```

ഞങ്ങൾ ഒരു ഉപയോക്തൃ സൗഹൃദ റാപ്പർ നൽകുന്നു, ഇത് നമുക്ക് അനുവദിക്കും:
- ജനറേഷന്റെ താപനില സജ്ജീകരിക്കുക
- Model object-ൽ നിന്ന് സൃഷ്ടിച്ച പേജ് പിടിക്കുക


```python
@guidance(stateless=True)
def html(
    lm,
    name: str | None = None,
    *,
    temperature: float = 0.0,
):
    return lm + capture(
        with_temperature(_gen_html(), temperature=temperature),
        name=name,
    )
```

ഞങ്ങൾ മോഡലിന് ഒരു പ്രോംപ്റ്റ് നൽകാം, പിന്നീട് ഒരു ജനറേഷൻ അഭ്യർത്ഥിക്കാം:


```python
lm = phi3_lm

lm += "Create a web page about your life story. Split your uplifting tale into multiple paragraphs with headings:\n"
lm += html(name="html_text", temperature=0.7)
```

അപ്പോൾ നാം ഔട്ട്‌പുട്ട് ഒരു ഫയലിലേക്ക് എഴുതാൻ കഴിയും:


```python
with open('./sample_page.html', 'w') as html_file:
    html_file.write(lm["html_text"])
```
