# Guidance ಬಗ್ಗೆ
Guidance ಒಂದು ಪ್ರಮಾಣಿತ ಮುಕ್ತ-ಮೂಲ ಪೈಥನ್ ಗ್ರಂಥಾಲಯವಾಗಿದ್ದು, ಯಾವುದೇ ಭಾಷಾ ಮಾದರಿಯ (LM) ಔಟ್‍ಪುಟ್‌ಗಳನ್ನು ನಿಯಂತ್ರಿಸಲು ಬಳಸಬಹುದು. ಒಂದು API ಕರೆ ಮೂಲಕ, ನೀವು (Python ನಲ್ಲಿ) ಮಾದರಿ ಪಾಲಿಸಬೇಕಾದ ನಿಖರ ಪ್ರೋಗ್ರಾಮ್ಯಾಟಿಕ್ ನಿರ್ಬಂಧ(ಗಳನ್ನು) ವಿವರಿಸಿ ಮತ್ತು JSON, Python, HTML, SQL ಅಥವಾ ಬಳಕೆದಾರರ ಅಗತ್ಯವಿರುವ ಯಾವುದೇ ರಚಿತ ಸ್ವರೂಪದಲ್ಲಿ ಸಂರಚಿತ ಔಟ್‍ಪುಟ್ ಅನ್ನು ರಚಿಸಬಹುದು.

Guidance ಪಾರಂಪರಿಕ ಪ್ರಾಂಪ್ಟಿಂಗ್ ತಂತ್ರಗಳಿಗಿಂತ ಭಿನ್ನವಾಗಿದೆ. ಇದು ಇನ್ಫರೆನ್ಸ್ ಲೇಯರ್‌ನಲ್ಲಿ ಪ್ರತಿ ಟೋಕನ್ ಅನ್ನು ಪ್ರತ್ಯೇಕವಾಗಿ ಮಾರ್ಗದರ್ಶನ ಮಾಡುವ ಮೂಲಕ ನಿರ್ಬಂಧಗಳನ್ನು ಜಾರಿಗೊಳಿಸುತ್ತಿದೆ, ಇದರಿಂದ ಉನ್ನತ ಗುಣಮಟ್ಟದ ಔಟ್‍ಪುಟ್‌ಗಳು ಉತ್ಪಾದಿಸುತ್ಇವೆ ಮತ್ತು ಅತ್ಯಂತ ರಚಿತ ಸಂದರ್ಭಗಳಲ್ಲಿ ಉಪಯೋಗಿಸಿದಾಗ ಖರ್ಚು ಮತ್ತು ವಿಳಂಬವನ್ನು 30–50% ವರೆಗೆ ಕಡಿಮೆ ಮಾಡುತ್ತದೆ.

Guidance ಬಗ್ಗೆ ಹೆಚ್ಚಿನ ಮಾಹಿತಿಗಾಗಿ, [GitHub‌ನಲ್ಲಿ ಇರುವ ಸಾರ್ವಜನಿಕ ರೆಪೊಸಿಟರಿ](https://github.com/guidance-ai/guidance) ಭೇಟಿ ನೀಡಿ ಅಥವಾ Microsoft Build ನಲ್ಲಿ ನಡೆದ [Guidance ಬ್ರೇಕೌಟ್ ಸೆಷನ್](https://www.youtube.com/watch?v=qXMNPVVlCMs) ವೀಕ್ಷಿಸಿ.


# ಸೆಟಪ್

1. Guidance ಅನ್ನು ಇನ್‌ಸ್ಟಾಲ್ ಮಾಡಿ `pip install guidance --pre`
2. Azure ನಲ್ಲಿ Phi 3.5 mini ಎಂಡ್‌ಪಾಯಿಂಟ್ ಅನ್ನು ಡೆಪ್ಲಾಯ್ ಮಾಡಲು https://ai.azure.com/explore/models/Phi-3.5-mini-instruct/version/2/registry/azureml ಗೆ ಹೋಗಿ ಮತ್ತು "Deploy" ಬಟನ್ ಕ್ಲಿಕ್ ಮಾಡಿ
3. ನಿಮ್ಮ ಎಂಡ್‌ಪಾಯಿಂಟ್‌ನ API ಕೀ ಅನ್ನು `AZURE_PHI3_KEY` ಎಂಬ ಪರಿಸರ ಚರದಲ್ಲಿ ಸಂಗ್ರಹಿಸಿ ಮತ್ತು URL ಅನ್ನು `AZURE_PHI3_URL` ಎಂಬ ಪರಿಸರ ಚರದಲ್ಲಿ ಸಂಗ್ರಹಿಸಿ


```python
from guidance import gen, select, regex, user, assistant, system, json
from guidance.models import AzureGuidance
from json import loads as load_json_str
import os

phi3_url = os.getenv("AZURE_PHI3_URL")
phi3_api_key = os.getenv("AZURE_PHI3_KEY")
phi3_lm = AzureGuidance(f"{phi3_url}/guidance#auth={phi3_api_key}")

# Or, load from HuggingFace to run locally
# from guidance.models import Transformers
# phi3_lm = Transformers("microsoft/Phi-3-mini-4k-instruct")
```

# ನಿಯಂತ್ರಣರಹಿತ ಉತ್ಪಾದನೆ
Text can be generated without any constraints using the `gen()` function. This is the same as using the model without Guidance.

## ಚಾಟ್ ಸ್ವರೂಪೀಕರಣ
Like many chat models, Phi-3 expects messages between a user and assistant in a specific format. Guidance supports Phi-3's chat template and will manage chat formatting for you. To create chat turns, put each portion of the conversation in a `with user()` or `with assistant()` block. A `with system()` block can be used to set the system message.


```python
lm = phi3_lm
with system():
    lm += "You are a helpful assistant. You have a cranky yet entertaining temperament."
with user():
    lm += "What is the capital of Australia?"
with assistant():
    lm += gen(temperature=0.8, max_tokens=100)
```

## ಟೋಕನ್ ಉಳಿತಾಯ
ಅತ್ಯಂತ ಸಂರಚಿತ ಸಂದರ್ಭಗಳಲ್ಲಿ, Guidance ಟೋಕನ್ಗಳನ್ನು ಬಿಟ್ಟುಹೋಗಿಸಿ ಕೇವಲ ಅಗತ್ಯವಿರುವ ಟೋಕನ್ಗಳನ್ನು ರಚಿಸಬಹುದು, ಇದರಿಂದ ಕಾರ್ಯಕ್ಷಮತೆ ಸುಧಾರಿಸುತ್ತದೆ, ದಕ್ಷತೆ ಹೆಚ್ಚುತ್ತದೆ ಮತ್ತು API ವೆಚ್ಚಗಳನ್ನು ಉಳಿತಾಯ ಮಾಡುತ್ತದೆ. ಉತ್ಪನ್ನವಾದ ಟೋಕನ್ಗಳು ಈ ನೋಟ್‌ಬುಕ್‌ನಲ್ಲಿ ಹೈಲೈಟ್ ಮಾಡಿದ ಹಿನ್ನೆಲೆಯೊಂದಿಗೆ ತೋರಿಸಲಾಗುತ್ತವೆ. ಒತ್ತಾಯಿತ ಟೋಕನ್ಗಳು ಹೈಲೈಟ್ ಇಲ್ಲದೆ ತೋರಿಸಲಾಗುತ್ತವೆ ಮತ್ತು ಇವುಗಳಿಗೆ ಇನ್‌ಪುಟ್ ಟೋಕನ್ಗಳಷ್ಟೇ ವೆಚ್ಚವಾಗುತ್ತವೆ; ಇನ್‌ಪುಟ್ ಟೋಕನ್ಗಳ ವೆಚ್ಚವನ್ನು ಔಟ್‌ಪುಟ್ ಟೋಕನ್ಗಳ ವೆಚ್ಚದ ಒಂದು-ಮೂರನೆಯಷ್ಟಾಗಿ ಅಂದಾಜಿಸಲಾಗಿದೆ.

*ಗಮನಿಸಿ:* ಮೊದಲ ಉದಾಹರಣೆ, ಅಡೆತಡೆ ರಹಿತ ಜನರೇಷನ್ ಹೊಂದಿದ್ದದ್ದರಿಂದ, ನಾವು ಯಾವುದೇ ನಿರ್ಬಂಧಗಳನ್ನು ಒದಗಿಸದ ಕಾರಣ ಯಾವ ಆರಂಭಿಕ ಟೋಕನ್ಗಳನ್ನೂ ಬಲವಂತವಾಗಿ ನಿಗದಿಪಡಿಸಲು ಸಾಧ್ಯವಾಗಲಿಲ್ಲ.


# Phi 3 ಪರವಾಗಿ ಮಾತನಾಡುವುದು 
Guidance ಬಳಸಿ, ನೀವು ಮಾದರಿಯ ಪ್ರತಿಕ್ರಿಯೆಗಳಲ್ಲಿ ಪಠ್ಯವನ್ನು ಸುಲಭವಾಗಿ ಸೇರಿಸಬಹುದು. ಇದು ನೀವು ಮಾದರಿಯ ಔಟ್‌ಪುಟ್ ಅನ್ನು ನಿರ್ದಿಷ್ಟ ದಿಕ್ಕಿನಲ್ಲಿ ಮಾರ್ಗದರ್ಶನ ಮಾಡಲು ಬಯಸಿದರೆ ಸಹಾಯಕವಾಗಬಹುದು.


```python
lm = phi3_lm
with user():
    lm += "What is the capital of Australia?"
with assistant():
    lm += "The capital of Australia is " + gen(temperature=0.8, max_tokens=50)
```

`ಆಸ್ಟ್ರೇಲಿಯಾದ ರಾಜಧಾನಿ` ಹೈಲೈಟ್ ಆಗಿಲ್ಲ ಏಕೆಂದರೆ ಸಹಾಯಕನ ಪ್ರತಿಕ್ರಿಯೆಯ ಆ ಭಾಗವು ಮಾರ್ಗದರ್ಶನದಿಂದ ಬಲವಂತವಾಗಿ ಸೇರಿಸಲಾಗಿತ್ತು.


# regex ಮೂಲಕ ನಿರ್ಬಂಧಿಸುವುದು

ಹಿಂದಿನ ಉದಾಹರಣೆಯಲ್ಲಿ, Phi 3 ಪ್ರಶ್ನೆಗೆ `Canberra` ಎಂದು ಉತ್ತರಿಸಿದ ನಂತರ ಅನುಸರಣಾ ವಿವರಗಳನ್ನು ನೀಡಿತು. ಮಾದರಿಯ ಔಟ್‌ಪುಟ್ ಅನ್ನು ನಿಖರವಾಗಿ ಒಂದೇ ಪದವಾಗಿರಿಸಲು, regex ಅನ್ನು ಬಳಸಬಹುದು.


```python
lm = phi3_lm
with user():
    lm += "What is the capital of Australia?"
with assistant():
    lm += "The capital of Australia is " + regex("[A-Z][a-z]+")
```

regex ನೊಂದಿಗೆ, ಕೇವಲ `Canberra` ಪದವೇ ರಚಿಸಲಾಗುತ್ತದೆ.


# ಬಹು ಆಯ್ಕೆಗಳಲ್ಲಿಂದ ಆಯ್ಕೆ ಮಾಡುವುದು
ಕೆಲವು ಸಾಧ್ಯವಾದ ಆಯ್ಕೆಗಳು ತಿಳಿದಿದ್ದರೆ, ಆಯ್ಕೆಗಳ ಪಟ್ಟಿಯಿಂದ ಮಾದರಿ ಆಯ್ಕೆ ಮಾಡುವಂತೆ ಮಾಡಲು ನೀವು `select()` ಕಾರ್ಯವನ್ನು ಬಳಸಬಹುದು.


```python
lm = phi3_lm
with user():
    lm += "What is the capital of Australia?"
with assistant():
    lm += "The capital of Australia is " + select(["Washington", "Canberra", "Sydney", "Melbourne"])
```

`select()` ಬಳಸದಾಗ, ಕೇವಲ ಟೋಕನ್ `Can` ಮಾತ್ರ ರಚಿಸಲಾಯಿತು. `Canberra` ಉತ್ತರವನ್ನು ಪೂರ್ಣಗೊಳಿಸಲು ಸಾಧ್ಯವಾಗಬಹುದಾದ ಏಕೈಕ ಆಯ್ಕೆಯಾಗಿದ್ದರಿಂದ, ಉಳಿದ ಟೋಕನ್‌ಗಳನ್ನು ಬಲವಶಾತ್ ಸೇರಿಸಲ್ಪಟ್ಟವು.


# ಚಿಂತನೆಗಳ ಸರಪಳಿ
ಚಿಂತನೆಗಳ ಸರಪಣಿ ಒಂದು ತಂತ್ರವಾಗಿದೆ, ಇದು ಸಮಸ್ಯೆಯನ್ನು ಹಂತ ಹಂತವಾಗಿ ಪ್ರಕ್ರಿಯೆಗೊಳಿಸಲು ಪ್ರೇರೇಪಿಸುವ ಮೂಲಕ ಮಾದರಿಯ ಫಲಿತಾಂಶದ ಗುಣಮಟ್ಟವನ್ನು ಸುಧಾರಿಸಲು ಸಹಾಯ ಮಾಡಬಹುದು. ಸಾಮಾನ್ಯವಾಗಿ, ಅಂತಿಮ ಉತ್ತರವನ್ನು ತಲುपಲು ಅನೇಕ ಪ್ರಾಂಪ್ಟ್ ತಿರುವುಗಳು ಅಗತ್ಯವಾಗುತ್ತವೆ. ಮೊದಲು, ಮಾದರಿಯನ್ನು ಹಂತ ಹಂತವಾಗಿ ಚಿಂತಿಸಲು ಸೂಚಿಸಿ. ಆಮೇಲೆ, ಅಂತಿಮ ಉತ್ತರವನ್ನು ನೀಡಿ ಎಂದು ಮಾದರಿಯನ್ನು ಮತ್ತೊಮ್ಮೆ ಪ್ರಾಂಪ್ಟ್ ಮಾಡಿ. ಸಾಮಾನ್ಯ ಚಾಟ್ ಇನ್ಫರೆನ್ಸ್ API ಗಳಲ್ಲಿ, ಇದಕ್ಕೆ 2 API ಕರೆಯುಗಳು ಬೇಕಾಗುತ್ತವೆ, ಮತ್ತು ಮಾದರಿ ರಚಿಸಿದ “ಚಿಂತನೆಗಳ ಸರಪಳಿ”ಗೆ ದ್ವಿಗುಣವಾಗಿ ಶುಲ್ಕ ವಿಧಿಸಲಾಗುತ್ತದೆ – ಒಂದು ಬಾರಿ ಮಾದರಿ ಅದನ್ನು ರಚಿಸಿದಾಗ ಔಟ್‌ಪುಟ್ ಟೋಕನ್‌ಗಳಾಗಿ, ಮತ್ತು ನಂತರ ಎರಡನೇ ಕರೆಗಾಗಿ ಇನ್‌ಪುಟ್ ಟೋಕನ್‌ಗಳಾಗಿ. Guidance ನೊಂದಿಗೆ,  ಸಂಪೂರ್ಣ ಬಹುಹಂತ ಪ್ರಕ್ರಿಯೆಯನ್ನು ಒಂದು API ಕರೆ ಭಾಗವಾಗಿ ಸಂಸ್ಕರಿಸಿ ಶುಲ್ಕ ವಿಧಿಸಲಾಗುತ್ತದೆ, ಇದರಿಂದ ವೆಚ್ಚ ಮತ್ತು ವಿಲಂಬ ಕಡಿಮೆಯಾಗುತ್ತದೆ.


```python
gsm8k_question = "Mark has a garden with flowers. He planted plants of three different colors in it. Ten of them are yellow, and there are 80% more of those in purple. There are only 25% as many green flowers as there are yellow and purple flowers. How many flowers does Mark have in his garden?"
lm = phi3_lm
with user():
    lm += gsm8k_question
with assistant():
    lm += "Let's think step by step. " + gen(temperature=0.8, max_tokens=500)
    # Prompt for the final answer, which should be a number. Store the output in an "answer" variable.
    lm += "\nTherefore, the final answer is: " + regex(r"\d+", name="answer")

print(f"Final answer: {lm['answer']}")
```

# JSON ರಚನೆ

JSON schema ಅಥವಾ pydantic ಮಾದರಿಯೊಂದಿಗೆ ಅನುಗುಣವಾಗಿರುವ JSON ಅನ್ನು ಖಚಿತಪಡಿಸಲು ಮಾರ್ಗದರ್ಶನವನ್ನು ಬಳಸಬಹುದು, ಉದಾಹರಣೆಗೆ ಇಲ್ಲಿ ತೋರಿಸುವ ಬಳಕೆದಾರ ಪ್ರೊಫೈಲ್ ಸ್ಕೀಮಾ.


```python
user_json_schema = load_json_str("""{
  "$schema": "http://json-schema.org/draft-07/schema#",
  "title": "User Profile",
  "type": "object",
  "properties": {
    "username": {
      "type": "string"
    },
    "age": {
      "type": "integer"
    },
    "email": {
      "type": "string"
    }
  },
  "additionalProperties": false
}
""")

lm = phi3_lm
with user():
    lm += "Generate a JSON object for a user profile. The profile should include a username, age, email, and nothing more."

with assistant():
    lm += json(schema=user_json_schema, temperature=1.0)
```

```python
from pydantic import BaseModel

class UserProfile(BaseModel):
    username: str
    age: int
    email: str


lm = phi3_lm
with user():
    lm += "Generate a JSON object for a user profile. The profile should include a username, age, email, and nothing more."

with assistant():
    lm += json(schema=UserProfile, temperature=1.0)
```

ನಂತರ ನಾವು ಈ ಫಂಕ್ಷನ್ ಅನ್ನು Azure AI ನಲ್ಲಿ Guidance-ಸक्षम ಮಾದರಿಯ ಮೇಲೆ ಚಾಲನೆ ಮಾಡುತ್ತೇವೆ.

*ಸೂಚನೆ:* ಇದು ಪೂರ್ಣ ವೈಶಿಷ್ಟ್ಯಗಳಿರುವ HTML ಜನರೇಟರ್ ಆಗುವುದಿಲ್ಲ; ಉದ್ದೇಶವೆಂದರೆ ನೀವು ನಿಮ್ಮ ವೈಯಕ್ತಿಕ ಅಗತ್ಯಗಳಿಗೆ ಅನುಗುಣವಾಗಿ ಸಂರಚಿತ ಫಲಿತಾಂಶವನ್ನು ಹೇಗೆ ರಚಿಸಬಹುದು ಎಂದು ತೋರಿಸುವುದು

ನಾವು Guidance ನಿಂದ ನಮಗೆ ಬೇಕಾದ 것을 ಆಮದು ಮಾಡುವುದರಿಂದ ಪ್ರಾರಂಭಿಸುತ್ತೇವೆ:


```python
from guidance import guidance
from guidance.library import (
    zero_or_more,
    any_char_but,
    select,
    capture,
    with_temperature,
)
from guidance.models import Model
```

HTML ವೆಬ್‌ಪೇಜ್‌ಗಳು ಅತ್ಯಂತ ಸಂರಚಿತವಾಗಿವೆ, ಮತ್ತು ನಾವು Guidance ಅನ್ನು ಉಪಯೋಗಿಸಿ ಆ ಪುಟದ ಭಾಗಗಳನ್ನು 'ಬಲವಂತವಾಗಿ' ನಿಯಂತ್ರಿಸುತ್ತೇವೆ.
When we explicitly require text from the model, we need to ensure it doesn't include anything which could be a tag - that is, we must exclude the '<' and '>' characters:


```python
@guidance(stateless=True)
def _gen_text(lm: Model):
    return lm + zero_or_more(any_char_but(["<", ">"]))
```

ನಂತರ ನಾವು ಈ ಫಂಕ್ಷನ್ ಅನ್ನು ಯಾವುದೇ HTML ಟ್ಯಾಗಿನ ಒಳಗೆ ಪಠ್ಯ ರಚಿಸಲು ಬಳಸಬಹುದು:


```python
@guidance(stateless=True)
def _gen_text_in_tag(lm: Model, tag: str):
    lm += f"<{tag}>"
    lm += _gen_text()
    lm += f"</{tag}>"
    return lm
```

ಈಗ, ನಾವು ಪುಟದ ಹೆಡರ್ ರಚಿಸೋಣ.
ಇದರ ಭಾಗವಾಗಿ, ನಮಗೆ ಪುಟದ ಶೀರ್ಷಿಕೆಯನ್ನು ರಚಿಸುವ ಅಗತ್ಯವಿದೆ:


```python
@guidance(stateless=True)
def _gen_header(lm: Model):
    lm += "<head>\n"
    lm += _gen_text_in_tag("title") + "\n"
    lm += "</head>\n"
    return lm
```

HTML ಪುಟದ ದೇಹವು ಶೀರ್ಷಿಕೆಗಳು ಮತ್ತು ಅನುಚ್ಛೇದಗಳಿಂದ ತುಂಬಲಾಗುತ್ತದೆ.
ಪ್ರತಿಯೊಂದನ್ನು ಮಾಡಲು ನಾವು ಪ್ರತ್ಯೇಕ ಕಾರ್ಯವನ್ನು ವ್ಯಾಖ್ಯಾನಿಸಬಹುದು:


```python
@guidance(stateless=True)
def _gen_heading(lm: Model):
    lm += select(
        options=[_gen_text_in_tag("h1"), _gen_text_in_tag("h2"), _gen_text_in_tag("h3")]
    )
    lm += "\n"
    return lm

@guidance(stateless=True)
def _gen_para(lm: Model):
    lm += _gen_text_in_tag("p")
    lm += "\n"
    return lm
```

ಈಗ, HTML ದೇಹವನ್ನು ನಿರ್ಧರಿಸುವ ಫಂಕ್ಷನ್.
ಇದು ಹಲವಾರು ಶೀರ್ಷಿಕೆಗಳು ಮತ್ತು ಅನুচ್ಛೇದಗಳನ್ನು ರಚಿಸಲು `select()` ಅನ್ನು `recurse=True` ನೊಂದಿಗೆ ಬಳಸುತ್ತದೆ:


```python
@guidance(stateless=True)
def _gen_body(lm: Model):
    lm += "<body>\n"
    lm += select(options=[_gen_heading(), _gen_para()], recurse=True)
    lm += "</body>\n"
    return lm
```

ಮುಂದೆ, ನಾವು ಸಂಪೂರ್ಣ HTML ಪುಟವನ್ನು ರಚಿಸುವ ಫಂಕ್ಷನ್‌ಗೆ ಬರುತ್ತೇವೆ.
ನಾವು HTML ಪ್ರಾರಂಭ ಟ್ಯಾಗ್ ಅನ್ನು ಸೇರಿಸುತ್ತೇವೆ, ನಂತರ ಶೀರ್ಷಿಕೆಯನ್ನು ರಚಿಸುತ್ತೇವೆ, ನಂತರ ದೇಹವನ್ನು ರಚಿಸುತ್ತೇವೆ, ಮತ್ತು ನಂತರ ಅಂತಿಮ HTML ಟ್ಯಾಗ್ ಅನ್ನು ಸೇರಿಸುತ್ತೇವೆ:


```python
@guidance(stateless=True)
def _gen_html(lm: Model):
    lm += "<html>\n"
    lm += _gen_header()
    lm += _gen_body()
    lm += "</html>\n"
    return lm
```

ನಾವು ಬಳಕೆದಾರ ಸ್ನೇಹಿ ರ್ಯಾಪರ್ ಅನ್ನು ಒದಗಿಸುತ್ತೇವೆ, ಇದು ನಮಗೆ ಅನುಮತಿಸುತ್ತದೆ:
- ಜನರೇಶನ್‌ನ ತಾಪಮಾನವನ್ನು ಹೊಂದಿಸುವುದು
- Model object ನಿಂದ ರಚಿಸಲಾದ ಪುಟವನ್ನು ಹಿಡಿಯುವುದು


```python
@guidance(stateless=True)
def html(
    lm,
    name: str | None = None,
    *,
    temperature: float = 0.0,
):
    return lm + capture(
        with_temperature(_gen_html(), temperature=temperature),
        name=name,
    )
```

ನಾವು ಮಾದರಿಗೆ ಪ್ರಾಂಪ್ಟ್ ಒದಗಿಸಬಹುದು, ಮತ್ತು ನಂತರ ಜನರೇಷನ್ ಅನ್ನು ವಿನಂತಿಸಬಹುದು:


```python
lm = phi3_lm

lm += "Create a web page about your life story. Split your uplifting tale into multiple paragraphs with headings:\n"
lm += html(name="html_text", temperature=0.7)
```

ನಂತರ ನಾವು ಔಟ್‌ಪುಟ್ ಅನ್ನು ಫೈಲ್‌ಗೆ ಬರೆಯಬಹುದು:


```python
with open('./sample_page.html', 'w') as html_file:
    html_file.write(lm["html_text"])
```
