## தேவையான நூலகங்களை இறக்குமதி செய்யவும்
Phi-4 மாதிரியை ஏற்றவும் பயன்படுத்தவும் தேவையான PyTorch மற்றும் transformers நூலகங்களை இறக்குமதி செய்க.


```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
```

## ரேண்டம் விதையை அமைக்கவும்
வெவ்வேறு ஓட்டங்களில் மறுபடியும் உருவாக்கக்கூடிய முடிவுகளை உறுதி செய்ய ரேண்டம் விதையை அமைக்கவும்.


```python
torch.random.manual_seed(0)
```

## Phi-4-mini-flash-reasoning மாதிரி மற்றும் Tokenizer ஐ ஏற்றவும்

Microsoft Phi-4-mini-flash-reasoningmodel மற்றும் அதனுடன் தொடர்புடைய tokenizer-ஐ Hugging Face இலிருந்து ஏற்றவும். மாதிரி வேகமான inference-க்கு CUDA இல் ஏற்றப்படும்.


```python
model_id = "microsoft/Phi-4-mini-flash-reasoning"
model = AutoModelForCausalLM.from_pretrained(
   model_id,
   device_map="cuda",
   torch_dtype="auto",
   trust_remote_code=True,
)
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

## உள்ளீட்டு செய்தியைத் தயாரிக்கவும்
மாதிரிக்கான உரையாடல் வார்ப்புருவைப் பயன்படுத்தி, ஒரு குவாட்ராடிக் சமன்பாடு அடங்கிய ஒரு உரையாடல் செய்தியை உருவாக்கவும்.


```python
messages = [{
   "role": "user",
   "content": "How to solve 3*x^2+4*x+5=1?"
}]   
inputs = tokenizer.apply_chat_template(
   messages,
   add_generation_prompt=True,
   return_dict=True,
   return_tensors="pt",
)
```

## பதில் உருவாக்கவும்
வெளியீட்டில் கட்டுப்படுத்தப்பட்ட சீரற்ற தன்மை பெற temperature மற்றும் top_p போன்ற குறிப்பிட்ட அளக்கூறுகளை பயன்படுத்தி மாதிரியிலிருந்து ஒரு பதிலை உருவாக்கவும்.


```python
outputs = model.generate(
   **inputs.to(model.device),
   max_new_tokens=32768,
   temperature=0.6,
   top_p=0.95,
   do_sample=True,
)
```

## வெளியீட்டை உரையாக மாற்றுதல்
உருவாக்கப்பட்ட டோக்கன் தொடர்களை மீண்டும் மனிதன் படிக்கக்கூடிய உரையாக மாற்றவும், அசல் உள்ளீட்டு டோக்கன்களை விலக்கி மாதிரியின் பதிலையே மட்டும் காட்ட.


```python
outputs = tokenizer.batch_decode(outputs[:, inputs["input_ids"].shape[-1]:])
```
